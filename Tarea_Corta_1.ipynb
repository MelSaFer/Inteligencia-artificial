```python
import numpy as np
import random

class Player:
    # Constructor
    def __init__(self, character):
        self.character = character

class Board:
    # Constructor
    def __init__(self, size):
        self.size = size
        self.horizontalLines = np.zeros((size, size-1), dtype=int)
        self.verticalLines = np.zeros((size-1, size), dtype=int)
        self.squares = np.full((size-1, size-1), '', dtype=str) # Represents every square

    """
    Method for ensuring that a move is valid
    Entries: 2 (x, y) where x is the row and y is the column
    """
    def isValidMove(self, position1, position2):
        if position1 == position2:
            print("Invalid move: Both positions are the same.")
            return False

        # Check if both positions are within the board boundaries
        if not (0 <= position1[0] < self.size and 0 <= position1[1] < self.size):
            print(f"Invalid move: Position {position1} is out of bounds.")
            return False
        if not (0 <= position2[0] < self.size and 0 <= position2[1] < self.size):
            print(f"Invalid move: Position {position2} is out of bounds.")
            return False

        # Check if the positions are adjacent
        if abs(position1[0] - position2[0]) + abs(position1[1] - position2[1]) != 1:
            print("Invalid move: Positions are not adjacent.")
            return False
        return True

    """
    Method for making a move in the board
    Entries:
            2 (x, y) where x is the row and y is the column
            the character of the player
    Output:
            -1: Invalid move
            0: No completed square
            1: Completed square
    """
    def makeMove(self, position1, position2, caracter):
        if not self.isValidMove(position1, position2):
            return -1

        position = None # variable that maps the x, y coords in the square matrix
        orientation = None
        # Check if the move is horizontal
        if position1[0] == position2[0]:
            orientation = True
            if position1[1] < position2[1]: # Left to right
                position = position1
            else:
                position = position2 # Rigth to left

            if self.horizontalLines[position[0]][position[1]] == 1:
                print("Invalid move: line already filled.")
                return -1
            self.horizontalLines[position[0]][position[1]] = 1

        # Check if the move is vertical
        elif position1[1] == position2[1]:
            orientation = False
            if position1[0] < position2[0]:
                position = position1 # Top to bottom
            else:
                position = position2 # Bottom to top
            if self.verticalLines[position[0]][position[1]] == 1:
                print("Invalid move: line already filled.")
                return -1
            self.verticalLines[position[0]][position[1]] = 1
        else:
            return -1
        return self.makeSquare(position, orientation, caracter)

    """
    Method for getting the available moves
    Output:
            A list of tuples with the available moves
    """
    def getMoves(self):
        moves = []

        for i in range(self.size):
            for j in range(self.size-1):
                if self.horizontalLines[i][j] == 0:
                    moves.append(((i, j), (i, j + 1)))  # Horizontal move
        for i in range(self.size-1):
            for j in range(self.size):
                if self.verticalLines[i][j] == 0:
                    moves.append(((i, j), (i + 1, j)))

        return moves

    """
    Method for checking if a move make a square
    Entries:
            2 (x, y) where x is the row and y is the column
            the character of the player
    Output:
            0: No completed square
            1: Completed square
    """
    def makeSquare(self, position, orientation, caracter):
        completed = 0
        row = position[0]
        column = position[1]

        # horizontal
        if orientation:
            # Check square above the current horizontal line
            if row > 0:
                if (self.horizontalLines[row-1][column] == 1 and  # Top line
                    self.verticalLines[row-1][column] == 1 and  # Left line
                    self.verticalLines[row-1][column+1] == 1):  # Right line
                    self.squares[row-1][column] = caracter  # Set the square to a default value (e.g., 'X')
                    completed += 1

            # Check square below the current horizontal line
            if row < self.size - 1:
                if (self.horizontalLines[row+1][column] == 1 and  # Bottom line
                    self.verticalLines[row][column] == 1 and  # Left line
                    self.verticalLines[row][column+1] == 1):  # Right line
                    self.squares[row][column] = caracter  # Set the square to a default value (e.g., 'X')
                    completed += 1

        else:  # Vertical line
            # Check square to the left of the current vertical line
            if column > 0:
                if (self.horizontalLines[row][column-1] == 1 and  # Top line
                    self.horizontalLines[row+1][column-1] == 1 and  # Bottom line
                    self.verticalLines[row][column-1] == 1):  # Left line
                    self.squares[row][column-1] = caracter  # Set the square to a default value (e.g., 'X')
                    completed += 1

            # Check square to the right of the current vertical line
            if column < self.size - 1:
                if (self.horizontalLines[row][column] == 1 and  # Top line
                    self.horizontalLines[row+1][column] == 1 and  # Bottom line
                    self.verticalLines[row][column+1] == 1):  # Right line
                    self.squares[row][column] = caracter  # Set the square to a default value (e.g., 'X')
                    completed += 1
        return completed

    """
    Method for checking if the game is over
    Output:
            True: The game is over
            False: The game is not over
    """
    def isTerminalState(self):
        return np.all(self.squares != '')

    """
    Method for printing the board
    """
    def print_board(self):
        board_size = len(self.squares) + 1  # Assuming squares is an (n-1) x (n-1) matrix

        # Print x-axis indices
        print("   ", end="")
        for j in range(board_size):
            print(f" {j}  ", end="")
        print()

        # Iterate over each row and column of the board
        for i in range(board_size):
            # Print y-axis index
            print(f"{i}  ", end="")

            # Print horizontal lines and intersection points
            for j in range(board_size):
                print("o", end="")  # Intersection point
                if j < board_size - 1:  # Ensure we stay within the horizontal lines range
                    if self.horizontalLines[i][j] == 1:
                        print("---", end="")  # Horizontal line
                    else:
                        print("   ", end="")  # Empty space between points
            print()  # Move to the next line for vertical lines or next row

            # Print vertical lines and squares (if not on the last row of points)
            if i < board_size - 1:  # Ensure we stay within the vertical lines range
                print("   ", end="")  # Align the vertical lines with the y-axis index
                for j in range(board_size):
                    if self.verticalLines[i][j] == 1:
                        print("| ", end="")  # Vertical line
                    else:
                        print("  ", end="")  # Empty space

                    # Print square contents (if not on the last column of points)
                    if j < board_size - 1:
                        if self.squares[i][j] != '':
                          print(f"{self.squares[i][j]} ", end="")  # Square content
                        else:
                          print("  ", end="")  # Empty space
                    else:
                        print("   ", end="")  # Align last column for vertical line
                print()  # Move to the next line for the next row of points

```

2. Desarrollo de la clase que maneja el min max

```python
class Minimax:
    # Constructor
    def __init__(self, max_depth, board, player1, player2):
        self.max_depth = max_depth
        self.board = board
        self.computer = player1  # Computer
        self.human = player2  # Other player

    """
    Method for evaluating the board (Euristica)
    Entries:
            The board to be evaluated
    Output:
            The evaluation of the board
    """
    def evaluate(self, board):
        computer_squares = np.sum(board.squares == self.computer.character)
        human_squares = np.sum(board.squares == self.human.character)
        return computer_squares - human_squares

    """
    Method for copying the board
    Entries:
            The board to be copied
    Output:
            The copied board
    """
    def copyBoard(self, board):
        newBoard = Board(board.size)
        newBoard.horizontalLines = np.copy(board.horizontalLines)
        newBoard.verticalLines = np.copy(board.verticalLines)
        newBoard.squares = np.copy(board.squares)
        return newBoard

    """
    Method for getting the best move
    Entries:
            isMaxPlayer: True if the current player is the computer, False otherwise
    Output:
            The best move
    """
    def bestMove(self, isMaxPlayer):
        best_move = None
        candidates = []

        # Initialize best_eval depending on whether we are maximizing or minimizing
        if isMaxPlayer:
            best_eval = float('-inf')
            character = self.computer.character
        else:
            best_eval = float('inf')
            character = self.human.character

        for move in self.board.getMoves():
            newBoard = self.copyBoard(self.board)
            score = newBoard.makeMove(*move, character)  # Make move with the correct character

            # The next move will switch roles: if currently maximizing, next will be minimizing and vice versa
            eval = self.minimax(newBoard, 0, not isMaxPlayer, score > 0)

            # Depending on whether we are maximizing or minimizing, update the best move and best evaluation
            if isMaxPlayer:
                if eval > best_eval:
                    best_eval = eval
                    candidates = [move]  # Reset candidates list if we find a better move
                elif eval == best_eval:
                    candidates.append(move)  # Add move to candidates if it's equally good
            else:  # Minimizing player's turn
                if eval < best_eval:
                    best_eval = eval
                    candidates = [move]  # Reset candidates list if we find a better move
                elif eval == best_eval:
                    candidates.append(move)  # Add move to candidates if it's equally good

        if candidates:
          # if there is more than one candidate
          if len(candidates) > 1:
              best_move = random.choice(candidates)  # selects one random move
          else:
              best_move = candidates[0]

        return best_move

    """
    Method for getting the minimax value
    Entries:
            board: The board to be evaluated
            depth: The depth of the search tree
            maximizing_player: True if the current player is the computer, False otherwise
            extra_turn: True if the current player made an extra turn, False otherwise
            isMaxPlayer: True if the current player is the computer, False otherwise
    Output:
            The minimax value
    """
    def minimax(self, board, depth, maximizing_player, extra_turn):
        if depth == self.max_depth or board.isTerminalState():
            return self.evaluate(board)

        if maximizing_player:
            max_eval = float('-inf')
            for move in board.getMoves():
                newBoard = self.copyBoard(board)
                score = newBoard.makeMove(*move, self.computer.character)
                eval = self.minimax(newBoard, depth + 1, True if score > 0 else False, score > 0)
                max_eval = max(max_eval, eval)
            return max_eval
        else:
            min_eval = float('inf')
            for move in board.getMoves():
                newBoard = self.copyBoard(board)
                score = newBoard.makeMove(*move, self.human.character)
                eval = self.minimax(newBoard, depth + 1, False if score > 0 else True, score > 0)
                min_eval = min(min_eval, eval)
            return min_eval
```

```python
import os
from IPython.display import clear_output
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

maxSize = 5
minSize = 2
minDepth = 1
maxDepth = 5

class Game:
    # Constructor
    def __init__(self):
        self.setBoard()
        self.players = [Player('O'), Player('X')]  # Player 1 is the computer (O), Player 2 is the human or computer (X)
        self.turn = 1  # 0 for player1 (computer), 1 for player2 (human)
        self.minimax = Minimax(self.setDepth(), self.board, self.players[0], self.players[1])
        self.minimax2 = None # for player 2 in case of computer vs computer
        self.movementCount = 0
        self.start_time = None
        self.end_time = None
        self.time_per_move = []

    """
    Method for changing the turn
    """
    def changeTurn(self):
        self.turn = 1 - self.turn  # Switch between 0 and 1
        return self.turn

    """
    Method for getting the turn
    Output:
            The turn
    """
    def getTurn(self):
        return self.turn

    """
    Method for setting the board
    """
    def setBoard(self):
        while True:
            try:
                size = int(input(f"Enter the size of the board (between {minSize} and {maxSize}): "))
                if size > maxSize or size < minSize:
                    print(f"Invalid input, enter a number between {minSize} and {maxSize}")
                    continue
                self.board = Board(size)
                break
            except ValueError:
                print("Invalid input, enter a valid number")

    """
    Method for setting the depth
    Output:
            The depth
    """
    def setDepth(self):
        while True:
            try:
                depth = int(input(f"Enter the depth of the search tree (between {minDepth} and {maxDepth}): "))
                if depth < minDepth or depth > maxDepth:
                    print(f"Invalid input, enter a number between {minDepth} and {maxDepth}")
                    continue
                return depth
            except ValueError:
                print("Invalid input, enter a valid number")

    """
    Method for incrementing the movement count
    """
    def incrementMovementCount(self):
        self.movementCount += 1

    """
    Method for getting the movement count
    Output:
            The movement count
    """
    def getMovementCount(self):
        return self.movementCount

    def computerTurn(self, minMax, isMaxPlayer):
        start_move_time = time.time()
        move = minMax.bestMove(isMaxPlayer)
        character = self.players[0].character if isMaxPlayer else self.players[1].character
        if move is not None:
            result = self.board.makeMove(move[0], move[1], character)
            self.incrementMovementCount()
            print("")  # Add space between moves
            print(f"{'Computer 1' if isMaxPlayer else 'Computer 2'} made move: {move}")
            print("---------------------------------------------------")
            if result == 0:  # If no square is completed, switch turn
                self.changeTurn()
            end_move_time = time.time()  # Fin del tiempo del turno
            self.time_per_move.append(end_move_time - start_move_time)  # Registra el tiempo de este movimiento
            return True
        return False


    def userTurn(self):
        start_move_time = time.time()
        while True:
            try:
                move1 = tuple(map(int, input("Enter first move position (row column): ").split()))
                move2 = tuple(map(int, input("Enter second move position (row column): ").split()))
                if not self.minimax.board.isValidMove(move1, move2):
                    print("Invalid move. Try again.")
                    continue
                result = self.minimax.board.makeMove(move1, move2, self.players[1].character)
                if result != -1:
                    print("")  # Add space between moves
                    self.incrementMovementCount()
                    print(f"You made move: {move1}, {move2}")
                    if result == 0:  # If no square is completed, switch turn
                        self.changeTurn()
                    break
            except ValueError:
                print("Invalid input. Enter row and column numbers separated by a space.")
        self.time_per_move.append(time.time() - start_move_time)
        return True

    def plot_times(self):
        df = pd.DataFrame(self.time_per_move, columns=['Time per Move'])
        df.plot(title="Time per Move", ylabel="Time (seconds)", xlabel="Move Number", kind='line')
        plt.show()
    """
    Method for playing the game
    """
    def play(self):
        print("Your character will be X and the computer will be O")
        self.start_time = time.time()
        while not self.minimax.board.isTerminalState():
            self.minimax.board.print_board()  # Print the current board state
            if self.getTurn() == 0:  # Computer's turn
                print("---------------------------------------------------")
                print("Computer's turn")
                result = self.computerTurn(self.minimax, False)
                if not result:
                    break

            else:  # Player's turn
                print("Your turn")
                result = self.userTurn()
                if not result:
                    break
        self.minimax.board.print_board()
        self.end_time = time.time()
        self.declare_winner()

    # computer vs computer
    def play_computer_vs_computer(self):
        print("-------------------------------")
        self.minimax2 = Minimax(self.setDepth(), self.board, self.players[1], self.players[0])
        self.start_time = time.time()
        while not self.board.isTerminalState():
            self.board.print_board()  # Print the current board state
            if self.getTurn() == 0:  # Computer 1's turn, maximizing
                print("Computer 1's turn")
                if not self.computerTurn(self.minimax, True):  # Pass True to maximize
                    break
            else:  # Computer 2's turn, minimizing
                print("Computer 2's turn")
                if not self.computerTurn(self.minimax2, False):  # Pass False to minimize
                    break
        self.board.print_board()
        self.end_time = time.time()
        self.declare_winner()


    """
    Method for declaring the winner and displaying game statistics
    """
    def declare_winner(self):
        print("\n========= Game Statistics =========\n")

        # Total time
        total_time = self.end_time - self.start_time
        print(f"Total Time: {total_time:.2f} seconds")

        avg_time_per_move = np.mean(self.time_per_move)
        print(f"Average Time per Move: {avg_time_per_move:.2f} seconds")

        # Count the squares for each player
        score_player1 = np.sum(self.board.squares == self.players[0].character)
        score_player2 = np.sum(self.board.squares == self.players[1].character)

        # Print the total moves of the game
        print(f"Total Moves: {self.getMovementCount()}")

        # Print the scores for each player
        print(f"Player 1 ({self.players[0].character}) Score: {score_player1}")
        print(f"Player 2 ({self.players[1].character}) Score: {score_player2}")

        print("\n---------------------------------------")

        # Determine and print the winner
        if score_player1 > score_player2:
            print(f"游끥 Player 1 ({self.players[0].character}) wins!")
        elif score_player2 > score_player1:
            print(f"游끥 Player 2 ({self.players[1].character}) wins!")
        else:
            print("游뱋 It's a tie!")

        print("\n=======================================\n")

        self.plot_times()
```

### *Clase Minimax*
Esta clase implementa el algoritmo Minimax para determinar el mejor movimiento en un juego basado en un tablero. Utiliza una funci칩n de evaluaci칩n heur칤stica para estimar la calidad de los estados del tablero y realiza una b칰squeda en el 치rbol de decisiones para seleccionar el movimiento 칩ptimo.
